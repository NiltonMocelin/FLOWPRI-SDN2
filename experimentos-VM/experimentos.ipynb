# Set up host: 

- Install docker in Alpine

- Etapas necessárias:
    -- Switch fica fora das VM, vamos criar várias bridges -> os switches possuem VTEPs para conectar o controlador e os hosts, e os outros switches
    -- Os controladores ficam de fora das VMs tbm, pois são apenas um serviço com diferentes interfaces e IPS
    -- Os hosts são os Qemu -> cada Qemu, precisa ter uma VTEP (porta/bridge) que conecta ao switch correto

- Configurar Alpine internet:
    -- subir interface: ifconfig eth0 up
    -- adicionar ip estatico ex: ip addr add 10.0.0.100/24 dev eth0
    -- adicionar rotas: ip route add 10.0.0.0/24 dev eth0
    -- adicinar default gateway: ip route add default via 10.0.0.1 dev eth0
    -- configurar dns: vi /etc/resolv.conf ; colocar nameserver 8.8.8.8 (outra linha) nameserver 8.8.4.4
    -- pronto
